```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)
```

```python
# create a dataframe with two classes
total_samples=1000
class_0_ratio=0.9
sample_0=int(total_samples*class_0_ratio)
sample_1=total_samples-sample_0
```

```python
sample_0,sample_1
```

```python
# create my dataframe with imbalance dataset

class_0=pd.DataFrame({
    'feature1':np.random.normal(loc=0, scale=1, size=sample_0),
    'feature2':np.random.normal(loc=0, scale=1, size=sample_0),
    'target':[0]*sample_0
})
class_1=pd.DataFrame({
    'feature1':np.random.normal(loc=2, scale=1, size=sample_1),
    'feature2':np.random.normal(loc=2, scale=1, size=sample_1),
    'target':[1]*sample_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.tail()
```

```python
df.head()
```

```python
df['target'].value_counts()
```

### Upsampling

```python
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_minority_upsampled=resample(df_minority,replace=True,
                                n_samples=len(df_majority),
                                random_state=42)
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```

### Down sampling

```python
import numpy as np
import pandas as pd

np.random.seed(123)

total_samples=1000
class_0_ratio=0.9
sample_0=int(total_samples*class_0_ratio)
sample_1=total_samples-sample_0

sample_0,sample_1

class_0=pd.DataFrame({
    'feature1':np.random.normal(loc=0, scale=1, size=sample_0),
    'feature2':np.random.normal(loc=0, scale=1, size=sample_0),
    'target':[0]*sample_0
})
class_1=pd.DataFrame({
    'feature1':np.random.normal(loc=2, scale=1, size=sample_1),
    'feature2':np.random.normal(loc=2, scale=1, size=sample_1),
    'target':[1]*sample_1
})

df=pd.concat([class_0,class_1]).reset_index(drop=True)

df.tail()

df.head()

df['target'].value_counts()

df_minority=df[df['target']==1]
df_majority=df[df['target']==0]

from sklearn.utils import resample
df_majority_downsampled=resample(df_majority,replace=True,
                                n_samples=len(df_minority),
                                random_state=42)

df_majority_downsampled.shape

df_majority_downsampled.head()

df_downsampled=pd.concat([df_minority,df_majority_downsampled])

df_downsampled['target'].value_counts()
```
